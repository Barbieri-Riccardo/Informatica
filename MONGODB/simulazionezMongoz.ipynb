1. I titoli dei film

```python
result = db.movies.find({},{"title":1, "_id":0})
df0 = pd.DataFrame(list(result))
df0
```

2. I film che nel titolo hanno la parola &quot;the&quot;

```python
result = db.movies.find({"title":{"$regex":"the"}},{"title":1, "_id":0})
df1 = pd.DataFrame(list(result))
df1
```

3. I film americani degli anni 20

```python
result = db.movies.find({"$and":[{"countries":"USA"},{"year":{"$gt":1919,"$lt":1930}}]},{"title":1, "_id":0})
df2 = pd.DataFrame(list(result))
df2
```

4. I film americani degli anni 30 e degli anni 50

```python
result=db.movies.find({'$and':[{'countries':"USA"}, {'$or': [{"year":{"$gt":1929, "$lt":1940}}, {"year":{"$gt":1949, "$lt":1960}}]}]})
df3 = pd.DataFrame(list(result))
df3
```

5. Tutti i film tranne quelli americani

```python
result = db.movies.find({"countries":{"$not":{"$regex":"USA"}}},{"title":1, "_id":0})
df4 = pd.DataFrame(list(result))
df4
```

6. I film che hanno un rating superiore a 4

```python
result = db.movies.find({"imdb.rating":{"$gt":4}},{"title":1, "_id":0})
df5 = pd.DataFrame(list(result))
df5
```

7. Il numero di film per ogni lingua. Ordinare in ordine decrescente a
partire della lingua più presente

```python
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }}]
)
df6 =pd.DataFrame(list(result))
df6
```

8. Le prime 3 lingue in cui sono stati girati i film

```python
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }},
                             {"$limit": 3}]
)
df7=pd.DataFrame(list(result))
df7
```

9. Le lingue in cui sono stati girati meno di 50 film

```python
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$group": {"_id" : "$languages", "total" : {"$sum" : 1}}},
                             {"$sort": { "total": -1 }},
                             {"$match": {"total" : {"$lte" : 50}}}]
)
df8 =pd.DataFrame(list(result))
df8
```

10. La durata media dei film italiani

```python
result=db.movies.aggregate( 
                            [{"$unwind": "$languages"},
                             {"$match" : {"languages" : "Italian"}},
                             {"$group": {"_id": "null", "media": {"$avg":"$runtime"} } }                            
                            ]
)
df9 =pd.DataFrame(list(result))
df9
```

11. La durata media dei film delle varie nazioni

```python
result=db.movies.aggregate( 
                            [{"$unwind": "$countries"},
                             {"$group": {"_id": "$countries", "media": {"$avg":"$runtime"} } }                            
                            ]
)
df10=pd.DataFrame(list(result))
df10
```
